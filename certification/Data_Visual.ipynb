# Histogram

```python
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt

df_canada = pd.read_csv("data_visual_1.csv")
df_canada['2013'].plot(kind='hist')

plt.title("Histogram of Imigration from several country")
plt.ylabel("Number of country")
plt.xlabel("Number of imigration")

plt.show()
```

```python
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt

import numpy as np

df_canada = pd.read_csv("data_visual_1.csv")

count, bin_edges = np.histogram(df_canada['2013'])

df_canada['2013'].plot(kind='hist', xticks=bin_edges)

plt.title("Histogram of Imigration from several country")
plt.ylabel("Number of country")
plt.xlabel("Number of imigration")

plt.show()
```

# Bar Chart

```python
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt

import numpy as np

df_canada = pd.read_csv("data_visual_1.csv")

years = list(map(str, range(1980, 2014)))

df_iceland = df_canada.loc[df_canada['OddName']=='Algeria', years]

df_iceland.plot(kind="bar")

plt.title['Algeria immigrate to canada from 1980 to 2013']
plt.xlabel['years']
plt.ylabel['Number of immigrant']

plt.show()
```
